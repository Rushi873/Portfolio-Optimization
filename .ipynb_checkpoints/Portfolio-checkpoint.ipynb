# Calculating different metrics

#### To understand the metrics for the portfolio, read the article, it explains the ratios in simple words and also what is their application and the optimal value "https://www.navexa.io/blog/8-portfolio-performance-metrics"

```python
def returns_calculation(price_data_frame, number_of_symbol, market_ticker):
    global random_weights, rebalance_weights, exp_returns, exp_vol, sharpe_ratio, beta, alpha, treynor_ratio
    
    # log accepted return
    log_return = np.log(1+ price_data_frame.pct_change())
    
    # Intialize random weights
    random_weights = np.round(np.array(np.random.random(number_of_symbol)), 5)
    
    # Rebalance to 1
    rebalance_weights = np.round(random_weights/np.sum(random_weights), 5)
    
    # Expected returns (for n number of days), len(price_data_frame) for number of columns ie day for which we are calculating
    exp_returns = np.sum((log_return.mean()* rebalance_weights)*(len(price_data_frame)))
    
    # Expected volatility (for n muber of days), standard deviation
    exp_vol = np.sqrt(
        np.dot( rebalance_weights.T, np.dot( log_return.cov() * len(price_data_frame), rebalance_weights))
    )
    
    # Sharpe ratio
    risk_free_rate = 0.071
    sharpe_ratio = (exp_returns - risk_free_rate) / exp_vol

    # Beta 
    market_returns = log_return[market_ticker]
    portfolio_returns = log_return.dot(rebalance_weights)
    
    cov_matrix = np.cov(portfolio_returns[1:], market_returns[1:])
    beta = cov_matrix[0, 1] / cov_matrix[1, 1]

    # Jensen's Alpha 
    expected_portfolio_return = risk_free_rate + beta * (exp_returns - risk_free_rate)
    alpha = exp_returns - expected_portfolio_return

    # Treynor Ratio 
    treynor_ratio = (exp_returns - risk_free_rate) / beta

    ## Sortino Ratio, alike sharpe ratio but considers the downside volatitility
    # downside_returns = log_return - risk_free_rate
    # downside_returns = np.minimum(downside_returns, 0)  # Take only negative returns
    # downside_volatility = np.std(downside_returns)
    # sortino_ratio = (exp_returns - risk_free_rate) / downside_volatility

    return random_weights, rebalance_weights, exp_returns, exp_vol, sharpe_ratio, beta, alpha, treynor_ratio
```

```python
# Runing an example simulation

# Initialize global variables
market_ticker = symbols[0]
random_weights = None
rebalance_weights = None
exp_returns = None
exp_vol = None
sharpe_ratio = None
beta = None
alpha = None
treynor_ratio = None
    
# Calculate portfolio metrics
random_weights, rebalance_weights, exp_returns, exp_vol, sharpe_ratio, beta, alpha, treynor_ratio = returns_calculation(price_data_frame, number_of_symbol, market_ticker)

# Convert weights to lists for DataFrame construction
random_weights = random_weights.tolist()
rebalance_weights = rebalance_weights.tolist()

# Weights dataframe
weights_df = pd.DataFrame(data={
    'random_weights' : random_weights,
    'rebalance_weights' : rebalance_weights
}, index=price_data_frame.columns)

print('')
print('='*80)
print('Portfolio weights:')
print('-'*80)
display(weights_df)
print('-'*80)

metrics_df = pd.DataFrame(data={
    "Expected Portfolio Returns" : exp_returns,
    "Expected Portfolio Volitility" : exp_vol,
    "Portfolio Sharpe Ratio" : sharpe_ratio,
    "Portfolio Beta" : beta,
    "Jensen's Alpha" : alpha, 
    "Treynor Ratio" : treynor_ratio
}, index=[0])

print('')
print('='*80)
print('Portfolio metrics:')
print('-'*80)
display(metrics_df)
print('-'*80)
```

# Applying Monte Carlo simulation

```python
# Number of simulation
num_simulation = 4000

#  Initialize array
all_weights = np.zeros((num_simulation, number_of_symbol))

# Return, vol and sharpe array
ret_arr = np.zeros(num_simulation)
vol_arr = np.zeros(num_simulation)
sharpe_arr = np.zeros(num_simulation)
beta_arr = np.zeros(num_simulation)
alpha_arr = np.zeros(num_simulation)
treynor_ratio_arr =  np.zeros(num_simulation)

# Simulation
for ind in range(num_simulation):
    random_weights, rebalance_weights, exp_returns, exp_vol, sharpe_ratio, beta, alpha, treynor_ratio = returns_calculation(price_data_frame, number_of_symbol, market_ticker)

    # appending to the array
    all_weights[ind, :] = rebalance_weights
    ret_arr[ind] = exp_returns
    vol_arr[ind] = exp_vol
    sharpe_arr[ind] = sharpe_ratio
    beta_arr[ind] = beta
    alpha_arr[ind] = alpha
    treynor_ratio_arr[ind] = treynor_ratio

# Combining them all together
simulation_data = [ret_arr, vol_arr, sharpe_arr, beta_arr, alpha_arr, treynor_ratio_arr, all_weights]

# Dataframe
simulation_df = pd.DataFrame(data=simulation_data).T
simulation_df.columns = [ "Returns", "Volatility", "Sharpe ratio", "Beta", "Alpha", "Treynor Ratio", "Portfolio Weights"]
simulation_df = simulation_df.infer_objects()

print('')
print('='*80)
print('Simulation Results:')
print('-'*80)
display(simulation_df.head(17))
print('-'*80)
```

# Optimization using Scipy

```python
# Using minimization as there is no maximazation in scipy
def get_metrics(weights: list, market_ticker) -> np.array:

    weights = np.array(weights)
    log_return = np.log(1+ price_data_frame.pct_change())
    ret = np.sum(log_return.mean()* weights) * len(price_data_frame)
    vol = np.sqrt(
        np.dot(weights.T, np.dot(log_return.cov()* len(price_data_frame), weights))
    )
    rfr = 0.0716
    sr = (ret - rfr)/vol
    mar_ret = log_return[market_ticker]
    portfolio_ret = log_return.dot(weights)
    cov_matrix = np.cov(portfolio_ret[1:], mar_ret[1:])
    beta = cov_matrix[0, 1] / cov_matrix[1, 1]
    exp_portfolio_ret = rfr + beta * (ret - rfr)
    alpha = ret - exp_portfolio_ret
    treynor_ratio = (ret - rfr) / beta

    return np.array([ret, vol, sr, beta, alpha, treynor_ratio])


def grab_transformed_sharpe(weights: list) -> np.array:
    return 1/(1 + get_metrics(weights, market_ticker)[2]) 

def grab_transformed_return(weights: list) -> np.array:
    return 1/(1 + get_metrics(weights, market_ticker)[0])

def grab_volitility(weights: list) -> np.array:
    return get_metrics(weights, market_ticker)[1]

def grab_transformed_alpha(weights: list) -> np.array:
    return 1/(1 + get_metrics(weights, market_ticker)[4])

def grab_transformed_treynor_ratio(weights: list) -> np.array:
    return 1/(1 + get_metrics(weights, market_ticker)[5])

def check_sum(weights: list) -> np.array:
    return np.sum(weights) -1
```

# For Sharpe ratio

```python
# Introducing boundaries
bounds = tuple((0,1) for symbol in range(number_of_symbol))
constraints = ({'type': 'eq', 'fun': check_sum})

init_guess = number_of_symbol*[1/number_of_symbol]

optimized_sharpe = sci_plt.minimize(
    grab_transformed_sharpe, 
    init_guess, 
    method='SLSQP',
    bounds=bounds,
    constraints=constraints
)

print('')
print('='*80)
print("Optimized sharpe ratio:")
print('-'*80)
print(optimized_sharpe)
print('-'*80)
print("We have obtained the Sharpe ratio in 1/(1 + sr) format")
```

```python
# for weights
optimized_metrics = get_metrics(weights=optimized_sharpe.x, market_ticker=market_ticker)

print('')
print('='*80)
print("Optimized weights:")
print('-'*80)
print(optimized_sharpe.x)
print('-'*80)
for i in range(0, len(symbols), 4):
    print(symbols[i:i+4])

print('\n \n')
print('='*80)
print("Optimized metrics:")
print('-'*80)
print(optimized_metrics)
print('-'*80)

print("[ Return , Volatility, Sharpe ratio, Beta, Alpha, Treynor ratio ]")
```
